```python
import pandas as pd
import sqlite3
import os

# Crear carpeta para la base de datos
if not os.path.exists('data'):
    os.makedirs('data')

# Datos de Turismo Honduras (2019-2023)
data = {
    'Anio': [2019, 2020, 2021, 2022, 2023] * 3,
    'Tipo_Ingreso': ['Crucero']*5 + ['Aéreo']*5 + ['Terrestre']*5,
    'Visitantes': [1500000, 380000, 620000, 1150000, 1400000, 
                   720000, 195000, 440000, 810000, 960000,
                   310000, 110000, 210000, 260000, 320000]
}

df = pd.DataFrame(data)

#  Guardar en SQLite
conn = sqlite3.connect('data/turismo_honduras.db')
df.to_sql('llegadas', conn, if_exists='replace', index=False)

# Probar una consulta rápida para verificar
df_check = pd.read_sql("SELECT * FROM llegadas WHERE Anio = 2023", conn)
conn.close()

print(" ¡Éxito! Base de datos 'turismo_honduras.db' creada en la carpeta /data")
df_check
```

```python
# Exportar la tabla de SQL a CSV para Power BI
conn = sqlite3.connect('data/turismo_honduras.db')
df_final = pd.read_sql("SELECT * FROM llegadas", conn)
df_final.to_csv('data/turismo_honduras_ready.csv', index=False)
conn.close()
print("CSV generado")
```
